```python
import pandas as pd
data=pd.read_csv("tripadvisor_hotel_reviews.csv")
data.head()
```

```python
df=data["Review"]
df.head()
```

```python
df.shape
```

```python
seq=[i.split() for i in df]
```

```python
seq
```

```python
!pip install gensim
```

```python
from gensim.models import Word2Vec
model=Word2Vec(seq,min_count=2)
```

```python
x=model.wv.vectors
```

```python
model.wv.index_to_key
```

```python
from sklearn.decomposition import PCA
pca=PCA(n_components=3)
cluster=pca.fit_transform(x)
```

```python
cluster[:0]
```

```python
cluster[:,1]
```

```python
cluster[:,2]
```

```python
import matplotlib.pyplot as plt
plt.scatter(cluster[:,0],cluster[:,1],cluster[:,2])
for i, word in enumerate(model.wv.index_to_key):
    plt.annotate(word,xy=(cluster[i,1],cluster[i,2]))
    plt.rcParams['font.sans-serif'] = ['Arial']
```

```python
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

# model.wv.vectors contains all the word embeddings
# Fit PCA to reduce dimensions from 100+ down to 2
pca = PCA(n_components=3)
result = pca.fit_transform(model.wv.vectors)
```

```python
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(10, 8))
# You MUST add projection='3d' to enable the 3rd axis
ax = fig.add_subplot(111, projection='3d')

# result[:, 0] is X, result[:, 1] is Y, result[:, 2] is Z
ax.scatter(result[:, 0], result[:, 1], result[:, 2])

words=model.wv.index_to_key
# Add word labels in 3D space
for i, word in enumerate(words):
    ax.text(result[i, 0], result[i, 1], result[i, 2], word)

ax.set_xlabel('PC 1')
ax.set_ylabel('PC 2')
ax.set_zlabel('PC 3')
plt.show()
```
