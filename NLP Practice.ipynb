```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(indep,dep,test_size=0.33,random_state=53)
```

```python
data=df
def sentiment(data,text):
    from nltk.sentiment.vader import SentimentIntensityAnalyzer
    sia=SentimentIntensityAnalyzer()
    score_dict=data[text].apply(lambda x: sia.polarity_scores(x))
    data["Neg"]=data[text].apply(lambda x: sia.polarity_scores(x)['neg'])
    data["Neutral"]=data[text].apply(lambda x: sia.polarity_scores(x)['neu'])
    data["Positive"]=data[text].apply(lambda x: sia.polarity_scores(x)['pos'])
    data["Compound"]=data[text].apply(lambda x: sia.polarity_scores(x)['compound'])
    data["Sentiment"]=data["Compound"].apply(lambda y: 'Positive' if y>0 else 'Negative' if y<0 else 'Neutral')
    return score_dict,data
```

```python
score_dict,data=sentiment(df,"Sentence")
```
