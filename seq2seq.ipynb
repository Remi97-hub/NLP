```python
def tokenize_padding(soseos_tokens,length,pad):
    tokenizer=Tokenizer(filters="")
    tokenizer.fit_on_texts(soseos_tokens)
    sequences=tokenizer.texts_to_sequences(soseos_tokens)

    maxlen = max(len(seq) for seq in sequences)+2

    padded_sequence=pad_sequences(sequences,maxlen=maxlen,padding=pad)
    return sequences,padded_sequence,tokenizer,maxlen
```

```python
def embedding_matrix(tokenizer,w2v_model):
    vocab_size=len(tokenizer.word_index)+1
    embedding_matrix=np.zeros((vocab_size,w2v_model.vector_size))

    for word,i in tokenizer.word_index.items():
        try:
            vector=w2v_model.wv[word]
            embedding_matrix[i]=vector
        except KeyError:
            pass
    return embedding_matrix
```

```python
eng_embdd_matrix=embedding_matrix(eng_tokenizer,eng_model)
tam_embdd_matrix=embedding_matrix(tam_tokenizer,tam_model)
```

```python
print(eng_embdd_matrix.shape)
print(tam_embdd_matrix.shape)
```

```python
def create_seq2seq_model(input_seq_length,output_seq_length,input_vocab_size,output_vocab_size,hidden_units,eng_embedding,tam_embedding):
    encoder_input=Input(shape=(input_seq_length,))
    encoder_embedding=Embedding(input_vocab_size,hidden_units,trainable=False,weights=[eng_embedding])(encoder_input)

    encoder_lstm = LSTM(hidden_units, return_state=True)
    _, encoder_state_h, encoder_state_c = encoder_lstm(encoder_embedding)
    
    
    encoder_states = [encoder_state_h, encoder_state_c]
    
    decoder_inputs=Input(shape=(output_seq_length,))
    decoder_embedding=Embedding(output_vocab_size,hidden_units,weights=[tam_embedding],trainable=False)(decoder_inputs)
    
    decoder_lstm=LSTM(hidden_units,return_sequences=True,return_state=True)
    decoder_outputs,_,_=decoder_lstm(decoder_embedding,initial_state=[encoder_state_h,encoder_state_c])
    decoder_dense=Dense(output_vocab_size,activation='softmax')
    decoder_outputs=decoder_dense(decoder_outputs)

    model=Model([encoder_input,decoder_inputs],decoder_outputs)
    return model
```
